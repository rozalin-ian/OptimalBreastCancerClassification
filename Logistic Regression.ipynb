# Data Preparation

## Importing Packages

```
# Importing the Libraries

import pandas as pd 
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline


import warnings
warnings.filterwarnings('ignore')
```

## Reading the Dataset

```
# Read the Breast Cancer Dataset

data = pd.read_csv('breast_cancer.csv')
```

```
# Explore the dataset

data.info()
data.head()
```

## Cleaning the Dataset

```
# Checking for missing values 

data.isnull().sum()
```

```
# Dropping the ID Column 

data = data.drop(['id'], axis = 1)
data.head()
```

```
# Check for duplicates 

data.duplicated().value_counts()
```

```
# Use Label Encoder to encode the diagnosis 
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Counting the number of benign and malignant
print(data['diagnosis'].value_counts())

# Transforming the diagnosis column
data['diagnosis'] = le.fit_transform(data['diagnosis'])

# Counting the number of 0s and 1s in the diagnosis
print(data['diagnosis'].value_counts())
```

```
# Handling Outliers

with_outliers = []
without_outliers = []

# Calculate the mean and standard deviation for each column
for column in data.columns:
    mean = data[column].mean()
    sd = data[column].std()
    
    # Define the boundaries
    lower_bound = mean - 3*sd
    upper_bound = mean + 3*sd
    
    # Find outliers
    outliers_below = data[column] < lower_bound
    outliers_above = data[column] > upper_bound

    #Replace outliers
   
    with_outliers_count = outliers_below.sum() + outliers_above.sum()
    with_outliers.append(column + " " + str(with_outliers_count))

    data[column][outliers_below] =  mean-3*sd
    data[column][outliers_above] =  mean+3*sd

    without_outliers_count = (data[column] < lower_bound).sum() + (data[column] > upper_bound).sum()
    without_outliers.append(column + " " + str(without_outliers_count))



print("With Outliers: ",with_outliers)
print("Without Outliers: ",without_outliers)
```

## Splitting the Dataset

```
# Splitting the Dataset into Training Set and Test Set

from sklearn.model_selection import train_test_split 

# Splitting the dataset into dependent and independent variables
X = data.iloc[:, 1:]
y = data.iloc[:, 0]

# Splitting the dataset into 20% test set and 80% training set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state = 7)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Feature Scaling

```
# Scale the X data using normalization 
from sklearn.preprocessing import MinMaxScaler

# Fit and transform the train data
min_max = MinMaxScaler()

X_train_scaled = min_max.fit_transform(X_train)
X_test_scaled = min_max.transform(X_test)
```

# Logistic Regression Model

```
from sklearn.linear_model import LogisticRegression
from sklearn.utils import resample


model_lr = LogisticRegression(penalty='l1', C=1, solver='liblinear', random_state=7)

# Fit and predict
model_lr.fit(X_train_scaled, y_train)
model_lr_predict = model_lr.predict(X_test_scaled)

print("Regularization Strength: ", (1/model_lr.C))
```

## Logistic Regression with Cross Validation to Evaluate Performance

```
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

kfold = KFold(n_splits=5, shuffle=True, random_state=7)

print('CV accuracy scores are', cross_val_score(model_lr, X_train_scaled, y_train, scoring='accuracy', cv=kfold, n_jobs=-1))
print('Average CV accuracy score', np.mean(cross_val_score(model_lr, X_train_scaled, y_train, scoring='accuracy', cv=kfold, n_jobs=-1)))
```

## Logistic Regression Confusion Matrix

```
# Confusion Matrix of Lasso Regression
from sklearn import metrics
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

matrix_best = metrics.confusion_matrix(y_test, model_lr_predict)

disp = metrics.ConfusionMatrixDisplay(confusion_matrix=matrix_best)
disp.plot()
```

## Classification Report of Logistic Regression

```
print(classification_report(y_test, model_lr_predict))
```

## Eliminated Features from the Model

```
# Get the names of the features
feature_names = data.drop('diagnosis', axis=1).columns

coefficients = model_lr.coef_[0]
zero_coeff_indices = np.where(coefficients == 0)[0]

removed_columns = feature_names[zero_coeff_indices]

print("Features with zero coefficients:", removed_columns)
```

## Remaining Features + Coefficients

```
coefficients = model_lr.coef_[0]

feature_names = data.drop('diagnosis', axis=1).columns

non_zero_coeff_indices = np.where(coefficients != 0)[0]

non_zero_feature_names = feature_names[non_zero_coeff_indices]

non_zero_coefficients = coefficients[non_zero_coeff_indices]

for feature, coefficient in zip(non_zero_feature_names, non_zero_coefficients):
    print("Feature:", feature, "Coefficient:", coefficient)
```

## Bias vs Variance

```
from mlxtend.evaluate import bias_variance_decomp

avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
    model_lr, X_train_scaled, y_train.values, X_test_scaled, y_test.values, 
    loss='0-1_loss', random_seed=7)

print('Average bias: %.3f' % avg_bias)
print('Average variance: %.3f' % avg_var)
```
