# Data Preparation

## Importing Packages

```
# Importing the Libraries

import pandas as pd 
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

## Reading the Dataset

```
# Read the Breast Cancer Dataset

data = pd.read_csv('breast_cancer.csv')
```

```
# Explore the dataset

data.info()
data.head()
```

## Cleaning the Dataset

```
# Checking for missing values 
data.isnull().sum()
```

```
# Dropping the ID Column 
data = data.drop(['id'], axis = 1)
data.head()
```

```
# Check for duplicates 
data.duplicated().value_counts()
```

```
# Use Label Encoder to encode the diagnosis 
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Counting the number of benign and malignant
print(data['diagnosis'].value_counts())

# Transforming the diagnosis column
data['diagnosis'] = le.fit_transform(data['diagnosis'])

# Counting the number of 0s and 1s in the diagnosis
print(data['diagnosis'].value_counts())
```

```
# Handling Outliers

with_outliers = []
without_outliers = []

# Calculate the mean and standard deviation for each column
for column in data.columns:
    mean = data[column].mean()
    sd = data[column].std()
    
    # Define the boundaries
    lower_bound = mean - 3*sd
    upper_bound = mean + 3*sd
    
    # Find outliers
    outliers_below = data[column] < lower_bound
    outliers_above = data[column] > upper_bound

    #Replace outliers
   
    with_outliers_count = outliers_below.sum() + outliers_above.sum()
    with_outliers.append(column + " " + str(with_outliers_count))

    data[column][outliers_below] =  mean-3*sd
    data[column][outliers_above] =  mean+3*sd

    without_outliers_count = (data[column] < lower_bound).sum() + (data[column] > upper_bound).sum()
    without_outliers.append(column + " " + str(without_outliers_count))



print("With Outliers: ",with_outliers)
print("Without Outliers: ",without_outliers)
```

## Splitting the Dataset

```
# Splitting the Dataset into Training Set and Test Set

from sklearn.model_selection import train_test_split 

# Splitting the dataset into dependent and independent variables
X = data.iloc[:, 1:]
y = data.iloc[:, 0]

# Splitting the dataset into 20% test set and 80% training set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state = 7)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```
# scale the X data using normalization
from sklearn.preprocessing import MinMaxScaler

min_max = MinMaxScaler()  
X_train_scaled = min_max.fit_transform(X_train)
X_test_scaled = min_max.fit_transform(X_test)
```

# K-Nearest Neighbour

```
from sklearn.neighbors import KNeighborsClassifier

#initializing this with no K value due to hyperparameter tuning later on
knn = KNeighborsClassifier()
```

## Hyperparameter Tuning

```
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.neighbors import KNeighborsClassifier

# Define the hyperparameter grid
param_grid = {'n_neighbors': range(1, 21)}

# Create a KFold object
kfold = KFold(n_splits=5, shuffle=True, random_state=7)

# Create a GridSearchCV object
grid_knn = GridSearchCV(KNeighborsClassifier(), param_grid, scoring='accuracy', cv=kfold, verbose=2)

# Fit the GridSearchCV object to the training data
grid_knn.fit(X_train, y_train)

# getting and printing best k value
best_k = grid_knn.best_params_['n_neighbors']
print("Best k value:", best_k)
```

## Fitting Hypertuned KNN Model

```
from sklearn.model_selection import cross_val_score

#best k value determined and fit into KNN model
best_knn = KNeighborsClassifier(n_neighbors=best_k)
best_knn.fit(X_test_scaled, y_test)

best_knn_predict = best_knn.predict(X_test_scaled)
```

## Confusion Matrix for Hypertuned KNN Model

```
#confusion matrix for the KNN model
from sklearn import metrics
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

matrix_best = metrics.confusion_matrix(y_test, best_knn_predict)

disp = metrics.ConfusionMatrixDisplay(confusion_matrix=matrix_best)
disp.plot()
```

## Classification Report

```
from sklearn.metrics import classification_report

print(classification_report(y_test, best_knn_predict))
```

## Accuracy Evaluation using K-Fold Cross Validation

```
#printing out the accuracy scores for each fold
cv_scores = cross_val_score(best_knn, X_train, y_train, cv=kfold, scoring='accuracy', n_jobs=-1)

#printing the accuracy scores
print('CV accuracy scores for each fold:', cv_scores)
print('Average CV accuracy score', np.mean(cv_scores))
```

## Bias and Variance 

```
from mlxtend.evaluate import bias_variance_decomp

avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
    best_knn, X_train_scaled, y_train.values, X_test_scaled, y_test.values, 
    loss='0-1_loss', random_seed=7)

print('Average bias: %.3f' % avg_bias)
print('Average variance: %.3f' % avg_var)
```
